```python
import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt('data.txt', skip_header=1)
```

```python
x = data[:,0]
y = data[:,1]
y_std = data[:,2]
```

```python
plt.errorbar(x, y, yerr=y_std, fmt='o', color='black', markersize=4, capsize=4)
plt.title('Plot of opt-out exercise data')
plt.ylabel('y')
plt.xlabel('x')
```

```python
def error(params):
    a, b, c = params

    f = a + b*x + c*x** 2

    err = (y-f)**2/y_std**2

    return np.sum(err)
```

```python
from scipy.optimize import minimize

guess_params = [0, 1, -1]

min_error = minimize(error, guess_params)
```

```python
plt.errorbar(x, y, yerr=y_std, fmt='o', color='black', markersize=4, capsize=4)
plt.title('Plot of opt-out exercise data')
plt.ylabel('y')
plt.xlabel('x')

sort_x = np.sort(x)

best_fit = min_error.x[0] + min_error.x[1]*sort_x + min_error.x[2]*sort_x**2
plt.plot(sort_x, best_fit, color='k')
```
